```python
import pandas as pd
import numpy as np

from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits.csv.base import create_csv_agent
```

```python
df = pd.read_csv('data/meta_data_sample.csv')
```

```python
var_names = df['Variable names'].values
var_names = var_names.tolist()
```

```python
tmp_file_path = 'data/meta_data_sample.csv'
agent_gpt = create_csv_agent(
        ChatOpenAI(temperature=0, 
                   model="gpt-3.5-turbo-0613",
                   request_timeout=120, 
                   openai_api_key='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'),
        tmp_file_path,
        verbose=True,
        agent_type=AgentType.OPENAI_FUNCTIONS,
    )
```

```python
def get_response_list(agent, var_names):
    response_list = []
    for var in var_names:
        question = "what does the variable" + var + " mean?"
        response = agent.run(question)
        response_list.append(response)
    return response_list  
```

```python
response_list = get_response_list(agent_gpt, var_names)
```

```python
df['agent_response'] = response_list
```

```python
df.to_csv('meta_data_with_response.csv', index=False)
```
